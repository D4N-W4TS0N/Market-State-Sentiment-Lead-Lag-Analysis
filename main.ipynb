```python
import os
import pandas as pd
import numpy as np
from dotenv import load_dotenv
from fredapi import Fred
import yfinance as yf

load_dotenv()
fred = Fred(api_key=os.environ["FRED_API_KEY"])
```

```python
icsa = fred.get_series("ICSA", observation_start='1999-01-01')
dgs2 = fred.get_series("DGS2", observation_start='19999-01-01')
dgs10 = fred.get_series("DGS10", observation_start='1999-01-01')
cpi = fred.get_series('CPIAUCSL', observation_start='1999-01-01')
vix = yf.download('^VIX', start='1999-01-01')['Close'].squeeze()
```

```python
print(icsa.head(), dgs2.head(), dgs10.head(), vix.head())
```

```python
idx = pd.date_range(start='2000-01-01', end='today', freq='B')

vix = vix.reindex(idx)

dgs2.index = pd.to_datetime(dgs2.index)
dgs10.index = pd.to_datetime(dgs10.index)
spread = (dgs10 - dgs2).reindex(idx)

icsa.index = pd.to_datetime(icsa.index)
icsa = icsa.reindex(idx, method='ffill')

cpi.index = pd.to_datetime(cpi.index)
cpiYoY = cpi.pct_change(12) * 100
cpiYoY = cpiYoY.reindex(idx).ffill()

df = pd.DataFrame({
    'icsa': icsa,
    'spread': spread,
    'vix': vix,
    'cpiYoY': cpiYoY
})

print(df.head())
print(df.tail())
```
